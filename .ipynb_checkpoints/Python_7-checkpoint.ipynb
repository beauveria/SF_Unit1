```python
#Определите, модель с каким item_id есть в статистике продаж purchase_df, но не учтена на складе
#(подсказка: подумайте, какой датафрейм должен быть "левым", а какой "правым", чтобы получить необходимые данные).
#Введите ответ в виде целого числа.
items_dict = {

    'item_id': [417283, 849734, 132223, 573943, 19475, 3294095, 382043, 302948, 100132, 312394], 

    'vendor': ['Samsung', 'LG', 'Apple', 'Apple', 'LG', 'Apple', 'Samsung', 'Samsung', 'LG', 'ZTE'],

    'stock_count': [54, 33, 122, 18, 102, 43, 77, 143, 60, 19]}
purchase_log = {

    'purchase_id': [101, 101, 101, 112, 121, 145, 145, 145, 145, 221],

    'item_id': [417283, 849734, 132223, 573943, 19475, 3294095, 382043, 302948, 103845, 100132], 

    'price': [13900, 5330, 38200, 49990, 9890, 33000, 67500, 34500, 89900, 11400]

}

items_df = pd.DataFrame(items_dict)
purchase_df = pd.DataFrame(purchase_log)

items_df.merge(purchase_df, how = 'right', on = 'item_id')
#Решите обратную задачу: модель с каким item_id есть на складе,
#но не имела ни одной продажи? Введите ответ в виде целого числа.
items_df.merge(purchase_df, how = 'left', on = 'item_id')
```

```python
#Сформируйте датафрейм merged, в котором в результате объединения purchase_df и items_df останутся модели, которые учтены на складе и имели продажи. 
#Сколько всего таких моделей?
merged = purchase_df.merge(items_df, on='item_id', how='inner')
display(merged)
```

```python
#Посчитайте объем выручки для каждой модели, которую можно получить, распродав все остатки на складе.
#Модель с каким item_id имеет максимальное значение выручки после распродажи остатков? Ответ дайте в виде целого числа.

merged = purchase_df.merge(items_df, on='item_id', how='outer')
merged['revenue']=(merged['stock_count']*merged['price'])

display(merged)
```

```python
#Посчитайте итоговую выручку из прошлого задания по всем моделям. Ответ дайте в виде целого числа.
import numpy as np
display(merged)
sum(merged['revenue'])
```

```python
import os
files = os.listdir('data')
files
```

```python
data = pd.DataFrame(columns = ['userId', 'movieId', 'rating', 'timestamp'])
#Напишите цикл, который собирает содержимое файлов папки data в единый датафрейм data.
#Сколько строк в датафрейме data
for filename in files:
    temp = pd.read_csv( os.path.join('data', filename), names = ['userId', 'movieId', 'rating', 'timestamp'])
    data = pd.concat([data, temp])
display(data)
```
